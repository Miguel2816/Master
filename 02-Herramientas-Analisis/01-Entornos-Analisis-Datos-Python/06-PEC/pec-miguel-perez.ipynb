<div style="background:#0099cc;padding:20px;color:#ffffff">
<p>Sustituye los valores nulos del campo edad por los valores medios obtenidos en funcin del sexo (es decir, sustituye los valores nulos de la edad de los hombres por la edad media de los hombres, y los valores nulos de la edad de las mujeres por la edad media de las mujeres). El proceso realizado comienza por encontrar las posiciones que se corresponde con:

* Sexo igual a female y valor nulo en el campo age
* Sexo igual a male y valor nulo en el campo age

Con las posiciones obtenidas, podemos hacer uso de loc con esas posiciones y la columna age para igualar los valores a las edades medias que obtenemos del dataframe anterior. Para el caso de sex igual a female, debemos obtener el valor [0,0], mientras que para el caso de sex igual a male, debemos obtener el valor [1,0].


```python
titanic_re.loc[(titanic_re['sex'] == 'female') & (titanic_re['age'].isnull()), 'age'] = mean_age.iloc[0,0]
titanic_re.loc[(titanic_re['sex'] == 'male') & (titanic_re['age'].isnull()), 'age'] = mean_age.iloc[1,0]
```

    A partir del dataset tips, crea una nueva columna (llmala "status") que tome el valor "alone" si en la comida en cuestin solo haba una persona, "couple" si haba 2, "group" si haba 3 o 4, y "crowd" si haba 5 o ms personas (recuerda que es el campo "size" el que contiene el nmero de personas de la comida).
</div>

```python
tips.head()
```

```python
tips['size'].value_counts(dropna=False) # COmprobamos los valores de la columna size
```

```python
def map_status(n):
    """
    Funcin diseada para recibir un valor entero de forma que si es igual a 1, devuelve alone, si es igual a 2 
    devuelve couple, si es igual a 3 o a 4 devuelve group, y si es igual a 5 o mayor, devuelve crowd.
    """
    if n==1:
        return 'alone'
    elif n==2:
        return 'couple'
    elif n==3 or n==4:
        return 'group'
    elif n > 4:
        return 'crowd'
```

```python
# Aplicamos la funcin anterior a la columna size con el uso de map
tips['status'] = tips['size'].map(map_status)
```

```python
tips['status'].value_counts(dropna=False)
```

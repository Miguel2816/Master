## Perceptrón multicapa

### Perceptrón multicapa con una capa Dense

[Keras](https://keras.io/) es una librería Python de alto nivel para trabajar con redes neuronales artificiales. Para ello utiliza paquetes de cálculo numérico y simbólico como Theano, TensorFlow o CNTK, lo que evita al usuario tener que programar los algoritmos básicos en lenguajes de bajo nivel como C, C++ o CUDA. Definir una red neuronal básica en Keras es así de sencillo:

```python
from keras.layers import Activation, Conv2D, Dense, Dropout, Flatten, MaxPooling2D
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from keras.models import Sequential

model = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])
```

A diferencia de Scikit-learn, esta es una librería específica para redes neuronales, pero también incluye algunas utilidades básicas para cargar datos como el dataset de [MNIST](http://yann.lecun.com/exdb/mnist/). Las imágenes de MNIST son de dígitos (de 0 a 9) manuscritos, en escala de grises (valores de 0 a 255) y de 28x28 (784) píxeles. El problema consiste en clasificar las imágenes en cada una de las 10 clases:

```python
from keras.datasets import mnist
from keras.utils import to_categorical

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
from keras.datasets import mnist
from keras.utils import to_categorical

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 784) / 255
x_test = x_test.reshape(10000, 784) / 255
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

Entrenar la red neuronal con los datos de train de MNIST y calcular las predicciones es también fácil:

```python
from keras.optimizers import SGD

model.compile(loss='categorical_crossentropy',
              optimizer=SGD(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=10, epochs=10,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

La accuracy en test no está mal, pero es muy mejorable. Prueba a crear una red neuronal con alguna capa densa más, entrénala durante más épocas y comprueba su accuracy en test:

Para ello, se genera a continuación el mismo modelo, pero añadiendo una capa densa y aumentando el número de épocas hasta 100.

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer=SGD(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=10, epochs=100,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Se puede observar como la accuracy aumenta con respecto al modelo anterior.
Repetimos el proceso con otra capa densa adicional, y en este caso 1000 épocas para ver como funciona el modelo.

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(784,)),
    Dense(32, activation='relu'),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer=SGD(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=10, epochs=1000,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Nuevamente el modelo ha vuelto a mejorar en la accuracy, pero muy poco con respecto al modelo anterior, y con una carga computacional mucho mayor debido a las 1000 épocas ejecutadas.

## Red convolucional

### Red convolucional con una capa Conv2D y una capa Dense

Como ya vimos en la teoría, las redes neuronales con capas convolucionales son especialmente potentes para procesamiento de imágenes. Estas capas se suelen colocar a la entrada de la red, seguidas de capas densas justo antes de la salida. Para poder aprovechar estas capas es necesario dejar las imágenes en su forma original 28x28:

```python
from keras.datasets import mnist
from keras.utils import to_categorical

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, 28, 28, 1) / 255
x_test = x_test.reshape(10000, 28, 28, 1) / 255
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
y_train = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)
```

Intenta crear una red neuronal con una capa [Conv2D](https://keras.io/layers/convolutional/#conv2d
) y una capa [Dense](https://keras.io/layers/core/#dense
) y repite el experimento anterior. Una indicación: entre la última capa convolucional y la primera densa debes usar una capa [Flatten](https://keras.io/layers/core/#flatten
):

```python
model = Sequential([
    Conv2D(filters=8, kernel_size=(3,3), activation='relu'),
    Flatten(),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer=SGD(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=10, epochs=100,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Se observa como al hacer uso de la capa convolucional se mejora considerablemente el modelo.
A continuación, se genera un nuevo modelo con una capa convolucional adicional y otra capa densa.

```python
model = Sequential([
    Conv2D(filters=8, kernel_size=(3,3), activation='relu'),
    Conv2D(filters=8, kernel_size=(3,3), activation='relu'),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer=SGD(),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=10, epochs=100,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Nuevamente se consigue mejorar la accuracy.

### Red convolucional avanzada

Todavía es posible hacerlo mejor. Intenta entrenar y hacer el scoring con una red neuronal más compleja, intercalando varias capas convolucionales y de pooling (pista: un buen modelo es la red LeNet-5). Puedes probar también otros métodos de optimización (Adam, RMSProp, etc.), regularizaciones (L1, L2, Dropout). Intenta conseguir el mejor resultado posible:

Se genera a continuación la red LeNet-5 mencionada en el enunciado, haciendo uso del método de optimización Adam y de regularización Dropout.

```python
model = Sequential([
    Conv2D(filters=6, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(filters=16, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(120, activation='relu'),
    Dense(84, activation='relu'),
    Dropout(0.25),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=10, epochs=100,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Se puede observar otra vez como la accuracy ha mejorado. 

Para intentar mejorar el modelo, se crea un función que genera que tendrá una arquitectura muy semejante a la red LeNet-5 para poder optimizar algunos parámetros.

```python
def make_model(output_dim, dense_layer_sizes, kernel_size=(3, 3), pool_size=(2, 2), dropout=0.25):
  
    model = Sequential()
    
    # Add pairs of convolutional-pooling hidden layers
    model.add(Conv2D(6, kernel_size, activation='relu'))
    model.add(MaxPooling2D(pool_size=pool_size))
    model.add(Conv2D(16, kernel_size, activation='relu'))
    model.add(MaxPooling2D(pool_size=pool_size))
    
    # Add the flatten layer before the dense ones
    model.add(Flatten())
    
    # Add the dense hidden layers
    for layer_size in dense_layer_sizes:
        model.add(Dense(layer_size, activation='relu'))

    
    # Add a dropout layer to avoid overfitting
    model.add(Dropout(dropout))
    
    # Add the output layer
    model.add(Dense(output_dim, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam',
                  metrics=['accuracy'])
    return model
```

Por falta de capacidad computacional, no se pueden probar muchos parámetros, por lo que se toman las siguientes consideraciones en cuanto a diferentes parámetros que se podrían optimizar:

* Épocas: se considera inviable intentar optmizar el número de épocas por falta de capacidad computacional y por el tiempo que llevaría su ejecución.
* Batch size: se va a probar con 10 y 32, siendo el número de muestras que se propagarán a través de la red durante la fase de entrenamiento.
* Optimizador: se entiende que el Adam funcionará significativamente mejor que el usado anteriormente, sgd, o que otras variantes del mismo. Se podría comprobar con el optimizador Nadam, pero no se considera que se vaya a encontrar una diferencia sustancial, y para evitar una carga computacional adicional se decide utilizar Adam.
* Función de activación: se considera que la función Relu es ampliamente utilizada y muy válida para este ejemplo, que consigue buenos resultados a pesar de su simplicidad y también posee la ventaja de que no satura.
* Dropout: se hicieron pruebas con diferentes parámetros y se observó que el valor 0.25 es un valor razonable con buenos resultados, por lo que se decide dejarlo así.
* Número de capas/neuronas: se va a intentar optimizar la arquitectura de las capas densas.

Como se ha comentado, estos son algunos de los parámetros que se pueden optimizar a la hora de generar el modelo. Simplemente se han eligido 2 por falta de capacidad computacional y por el tiempo de ejecuión a la hora de realizar pruebas, aunque es obvio que determinadas pruebas con alguno de estos parámetros podrían ayudar a conseguir un modelo con una mayor capacidad predictiva.

```python
classifier = KerasClassifier(make_model, output_dim=10, epochs=100)
search_space = {'dense_layer_sizes': [(120, 84), (120, 84, 58)], 
                'batch_size': [10,32]}
estimator = GridSearchCV(classifier, search_space, cv=3)
estimator.fit(x_train, y_train)
print('Best params: ' + str(estimator.best_params_))
print('Best score: ' + str(estimator.best_score_))
```

Se observa que el modelo con un mayor scoring se consigue con un batch size de 32 y con tres capas ocultas densas de 120, 85 y 58 unidades cada una.

A continuación se imprime el accuracy para poder compararlo con los modelos anteriores.

```python
predictions = estimator.predict(x_test)
print('Test accuracy: {}'.format(accuracy_score(y_test, predictions)))
```

Parece que los resultados de este modelo son ligeramente inferiores a los del anterior, cuyo batch size era igual a 10 y su arquitectura era de 2 capas ocultas con 120 y 84 unidades cada uno, pero esta era una de la combinaciones que se generaron en el gridsearch. Se imprimen a continuación los resultados de todas las combinaciones:

```python
estimator.cv_results_
```

Se observa como todos los valores son similares, pero hay una diferencia, y es que este entrenamiento se ha hecho siguiendo la metodología cross validation. Se genera a continuación el modelo con la arquitectura comentada, pero de igual forma que se había generado anteriormente:

```python
model = Sequential([
    Conv2D(filters=6, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(filters=16, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(120, activation='relu'),
    Dense(84, activation='relu'),
    Dense(58, activation='relu'),
    Dropout(0.25),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=32, epochs=100,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Parece que en este caso si que obtenemos un accuracy superior a los anteriores.

A continuacción se hacen pruebas aumentando el número de épocas:

```python
model = Sequential([
    Conv2D(filters=6, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Conv2D(filters=16, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Flatten(),
    Dense(120, activation='relu'),
    Dense(84, activation='relu'),
    Dense(58, activation='relu'),
    Dropout(0.25),
    Dense(10, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=32, epochs=200,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Finalmente, al ser un problema tan ampliamente resuelto en la comunidad, se encuentran multitud de blogs y artículos sobre el mismo, explicando diferentes arquitecturas. Tras una lectura de muchos de ellos, he decidido probar la siguiente arquitectura, pero debido a las dimensiones de la red, se decide únicamente ejecutar 10 épocas por falta de capacidad computacional.

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(5,5), activation='relu'),
    Conv2D(filters=32, kernel_size=(5,5), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.5),
    
    Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    Dropout(0.5),

    Flatten(),
    Dense(8192, activation='relu'),
    Dropout(0.5),
    Dense(2048, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])


model.compile(optimizer='adam',
              loss="categorical_crossentropy",
              metrics=["accuracy"])

history = model.fit(x_train, y_train, batch_size=32, epochs=10,
                    verbose=1, validation_split=0.1)

score = model.evaluate(x_test, y_test_cat, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Este modelo con únicamente 10 épocas ya consigue la test loss más baja de todos los entrenados anteriormente y una accuracy de las más altas, igual al modelo generado con la arquitectura LeNet-5, pero con la diferencia que este modelo ha sido entrenado con 10 épocas, y el de la arquitectura LeNet-5 con 100, por lo que es presumible que aumentando el número de épocas se consigan resultados mucho mejores que los obtenidos anteriormente.

<div style="font-size: 40px;text-align: center;height:50px;padding:10px;margin:0 0 10px 0;">Modos de trabajo</div>

Habrás observado que siempre hay una celda seleccionada que muestra un borde verde o azul: Este color indica el modo de trabajo en el que estamos. Si es verde estamos en modo de edición, y si es azul estamos en modo comando. Puedes entrar en modo de edición haciendo clic en el interior de una celda (en el rectángulo horizontal alargado donde podemos escribir -véase la siguiente figura-). Y puedes entrar en modo comando haciendo clic en cualquier punto de la celda que no sea el rectángulo anterior, por ejemplo en el extremo izquierdo de la celda, donde se muestra un área blanca más amplia:

Cuando estamos en modo de edición estamos modificando el contenido de la celda. Por ejemplo, estando en este modo de trabajo, seleccionar un texto con el ratón y presionar Control-C lo va a copiar al portapapeles de Windows (texto que podremos pegar en cualquier otra parte con Control-V, como sabes).

Por el contrario, cuando estamos en modo comando estamos trabajando con la celda, no con su contenido. Por ejemplo, estando en modo comando (recuerda, fondo azul) si presionamos la tecla C estamos copiando la celda en cuestión a la memoria. Ahora podemos hacer clic en cualquier otra celda para seleccionarla y presionar la tecla V: se pegará la celda que habíamos copiado debajo de la celda seleccionada.

Otra forma sencilla de pasar de un modo a otro es la siguiente: si estás en modo de edición (fondo verde) pulsa la tecla escape para "subir" al modo comando (fondo azul). Y si estás en modo comando, presiona la tecla "Intro" para "bajar" al modo de edición.

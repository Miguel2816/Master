Revisemos algunas de las opciones que tenemos disponibles en el menú de opciones que encontramos en todo cuaderno Jupyter:

###### File

* **New Notebook**: Nos permite crear un nuevo cuaderno Jupyter
* **Open**: Nos permite abrir un cuaderno Jupyter existente

###### Insert

* **Insert Cell Above**: Inserta una celda encima de la seleccionada (accesible también con el atajo de teclado A si estamos en modo de comando)
* **Insert Cell Below**: Inserta una celda debajo de la seleccionada (mismo comentario con la tecla B)

###### Cell

* **Run All**: Ejecuta todas las celdas sin reinicializar el cuaderno
* **Run All Above**: Ejecuta las celdas que haya por encima de la celda seleccionada (sin incluir ésta)
* **Run All Below**: Ejecuta la celda seleccionada y las que haya por debajo

###### Kernel

* **Interrupt**: Interrumpe la ejecución del cuaderno
* **Restart**: Reinicializa el cuaderno pero no borra los mensajes de salida de cada celda que se haya ejecutado
* **Restart & Clear Output**: Reinicializa el cuaderno y borra los mensajes de salida de cada celda que se haya ejecutado
* **Restart & Run All**: Reinicializa el cuaderno y ejecuta todas las celdas del mismo

Comenzamos creando la funcin aplicable a un simple escalar:

Y la aplicamos al array indicado:

Aplicamos la funcin np.square al mismo array:

```python
n = 1000000
```

```python
%%time
for i in range(n):
    cuadrado(a)
```

```python
%%time
for i in range(n):
    np.square(a)
```

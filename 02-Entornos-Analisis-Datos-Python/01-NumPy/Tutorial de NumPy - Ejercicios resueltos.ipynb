<div>Crea una función vectorizada ("cuadrado") que eleve al cuadrado los valores de un array y pruébala con el array:</div>
<div>a = np.arange(10)</div>
<div>Aplica la función np.square() al mismo array</div>
<div>Compara los tiempos de ejecución de cuadrado y de np.square aplicando estas funciones un millón de veces al array a</div>

Comenzamos creando la función aplicable a un simple escalar:

```python
def cuadrado_sencillo(n):
    return n*n
```

Ahora creamos la versión vectorizada:

```python
cuadrado = np.vectorize(cuadrado_sencillo)
```

Y la aplicamos al array indicado:

```python
a = np.arange(10)
a
```

```python
cuadrado(a)
```

Aplicamos la función np.square al mismo array:

```python
np.square(a)
```

Comprobamos que el resultado es el mismo. ¿Los tiempos de ejecución también lo son?

```python
n = 1000000
```

```python
%%time
for i in range(n):
    cuadrado(a)
```

```python
%%time
for i in range(n):
    np.square(a)
```

La función np.square es del orden de 20 veces más rápida que nuestra función vectorizada.

```python
np.where(a % 2 != 1, a, -1)
```

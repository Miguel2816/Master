# EJERCICIOS

# 1. Cargar la red social del club de kárate (1 punto)

Pista: buscad entre los generadores de NetworkX.


```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import community

# Se carga la red social del club de kárate
G = nx.karate_club_graph()
```

# 2. Dibujar la red social del club de kárate  (1 punto)

```python
# Se visualiza la red social. Se hace uso de una visualización circular que ayuda al entendimiento de la red. 
nx.draw_circular(G, with_labels=True)
plt.show()
```

# 3. Datos básicos (1 punto)
1. Número de nodos y número de aristas.
2. Densidad.
3. Diámetro.
4. Grado máximo, mínimo y promedio.

##### 1. Número de nodos y número de aristas

```python
print("El número de nodos es {}.".format(len(list(G.nodes()))))
print("El número de aristas es {}.".format(len(list(G.edges()))))
```

##### 2. Densidad

La densidad permite interpretar si hay muchas o pocas conexiones.

```python
print("La densidad de la red es {}.".format(nx.density(G)))
```

##### 3. Diámetro

Es la distancia geodésica más larga entre todos los posibles pares de nodos.

```python
print("El diámetro de la red es {}.".format(nx.diameter(G)))
```

###### 4. Grado máximo, mínimo y promedio

```python
# Se genera una lista vacía
degree_list = []

# Se añade a la lista el grado de cada nodo
for _,j in nx.degree(G):
    degree_list.append(j)
    
# Se imprime por pantalla los resultados
print("El grado máximo de la red social Zachary es {}".format(max(degree_list)))
print("El grado mínimo de la red social Zachary es {}".format(min(degree_list)))
print("El grado promedio de la red social Zachary es {}".format(sum(degree_list)/len(degree_list)))
```

El grado solo puede ser un número entero, por lo que si redondearamos el promedio a un entero, se podría decir que el grado promedio es 5.

# 4. Identifica al instructor y al administrador. (1 punto)

Son los dos nodos de mayor grado.

```python
# Ordenar la lista con orden descendiente
degree_list_sorted = sorted(degree_list, reverse = True)
print("El mayor grado de un nodo es {}, el segundo mayor es {} y el tercer mayor es {}.".format(degree_list_sorted[0],
                                                                                                degree_list_sorted[1],
                                                                                                degree_list_sorted[2]))
```

Se comprueba que los nodos que tienen un mayor grado son los que tienen un grado 16 y 17.

```python
print("Los nodos correspondientes al instructor y al administrador \
son el nodo {} y el nodo {}.".format(degree_list.index(17), degree_list.index(16)))
```

# 5. Vuelve a dibujar el grafo pero, esta vez, resaltando los nodos del instructor y del administrador. (1 punto)

(Es suficiente con que tengan un color distinto, por ejemplo)

```python
# Se genera una lista para mapear los colores. Los nodos 0 y 33 tendrán color azul y el resto verde.
color_map = []
for node in G:
    if node == 0 or node == 33:
        color_map.append('blue')
    else: 
        color_map.append('green')  

# Se dibuja el resultado para comprobarlo
nx.draw_circular(G, node_color=color_map, with_labels=True)
plt.show()
```

# 6. Calcula la centralidad de cercanía (closeness) y de intermediación (betweenness). (1 punto)

```python
# Se obtiene la centralidad de cercanía
nx.closeness_centrality(G)
```

```python
# Se obtiene la centralidad de intermediación
nx.betweenness_centrality(G)
```

# 7. Estudia el prestigio en la red mediante PageRank. (1 punto)

```python
# Se genera los resultados de aplicar el algoritmo PageRank a la red
nx.pagerank(G)
```

# 8. Obtén el coeficiente de agrupamiento medio y el número de triángulos completos (1 punto)

```python
# Coeficiente de agrupamiento medio
print("El coeficiente de agrupamiento medio es {}.".format(nx.average_clustering(G)))
```

Para el cálculo de triángulos completos, se calcula los triángulos comletos de cada nodo, se suman y se divide por 3, ya que cada triángulo se ha contado 3 veces.

```python
# Número de triángulos completos
print("El número de triángulos completos es {}.".format(int(sum(nx.triangles(G).values())/3)))
```

# 9. Obtén las comunidades según el algoritmo de Girvan-Newmann (1 punto)

```python
comp = community.girvan_newman(G)

print("La primera comunidad está formada por los nodos: {}.".format(tuple(sorted(c) for c in next(comp))[0]))
print("La segunda comunidad está formada por los nodos: {}.".format(tuple(sorted(c) for c in next(comp))[1]))
```

Se comienza analizando los resultados del apartado 6, donde se ha calculado la centralidad de cercanía y de intermediación, y es necesario entender estos conceptos. Mientras que la centralidad de intermediación mide si el nodo se encuentra en muchos caminos críticos, es decir, es un gestor de información, la centralidad de cercanía mide como de cerca está el nodo de los otros nodos, de forma que cuanto mayor sea, mayor será la capacidad de interacción del nodo. 

En esta red, se observa que la diferencia entre centralidades de cercanía de todos los nodos no es muy elevada, y esto se debe a que los nodos de la red están conectados entre sí en cierta manera, lo que hace que todos ellos, con pocos saltos, puedan conectarse a todos los nodos de la red, algo que se demuestra con el bajo valor resultante de calcular el diámetro de la red. También nos reafirma en esta hipótesis el valor del coeficiente de agrupación medio, que nos indica en que grado los nodos están conectados entre sí, y este valor es relativamente alto.

En cambio, la centralidad de intermediación es todo lo contrario, y nos permite observar dos nodos que destacan sobre el resto, y que son los correspondientes al instructor y al administrador, y hacen de puente entre conexiones de varios nodos, lo que hace que su centralidad de intermediación sea alta. Esta conclusión también se puede ver reforzada por el número de triángulos completos, que es 45, un valor bajo para la cantidad de nodos que hay, por lo que hay muchos nodos que no estan conectados entre sí y que gracias a los nodos 0 y 33 se pueden conectar a muchos otros.

El algortimo pagerank nos ayuda a analizar como son los enlaces, y es parecido a la centralidad de vector propio, por lo tanto, lo que nos indica no es tanto cuanto enlaces tiene un nodo, sino si estos enlaces son con nodos importantes. Nuevamente destacan sobre el resto los nodos 0 y 33 porque están conectados a muchos nodos, lo que les da importancia, pero como la gran mayoría de nodos está conectado a uno de los dos, ninguno tiene un valor igual a 0 o excesivamente bajo, ya que están todos conectados a un nodo importante.

Finalmente, el algoritmo de Girvan-Newmann nos permite obtener las dos comunidades en función de como se relacionan los nodos con los que representan al administrador y al instructor, pero también en función de como lo hacen con el resto.
